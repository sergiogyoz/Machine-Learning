# 4.1 Linear Regression

Linear regression model:
$\hat{y}=\theta_0+\theta_1 x_1 + \cdots + \theta_n x_n$, where

* $\hat{y}$: predicted value
* $n$: number of features/attributes
* $x_i$: $i$-th feature value
* $\theta_i$: $i$-th model parameter

Alternatively,
$\hat{y} = \mathbf{\theta}^T \mathbf{x}$.

For this to be the same, we need

* $\mathbf{\theta}:=[\theta_0, \theta_1,\cdots,\theta_n]^T$ and
* $\mathbf{x}:=[1, x_1,\cdots,x_n]^T$,

so that

$\mathbf{\theta}^T \mathbf{x}
=\theta_0+\theta_1 x_1 + \cdots + \theta_n x_n.$

That is, we insert constant 1 as the 0-th feature in the example. It's very common to do so, and we'll see how soon.




Typical performance measure for regression was the MSE:

$MSE(X,Y,\theta) = \frac{1}{m} \sum_i (\theta^T x^{(i)}-y^{(i)})^2$
where $m$ is the number of examples, and $x^{(i)}$ and $y^{(i)}$ means $i$-th example and $i$-th true label. 


## The Normal Equation

Q. What is the goal of training, mathmatically speaking?

A. $\min_{\theta}\;\;Cost(X,y,\theta)$

Lin Reg:
$\min_{\theta}\;MSE(X,Y,\theta) = \frac{1}{m} \sum_i (\theta^T x^{(i)}-y^{(i)})^2$


Can we solve this? 
Linear algebra says
$\hat{\theta} = (X^T X)^{-1} X^T y$, where

$\hat{\theta}$ is the solution (or optimum) of the min problem.

Let's check if this is true.

```python
import numpy as np

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
#save_fig("generated_data_plot")
plt.show()
```

$\hat{\theta} = (X^T X)^{-1} X^T y$

```python
X_b = np.c_[np.ones((100, 1)), X]  # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

There are several ways to compute $\hat{\theta} = (X^T X)^{-1} X^T y$ numerically.

The `LinearRegression` class is based on the `scipy.linalg.lstsq()` function (the name stands for "least squares"), which you could call directly:

```python
theta_best_svd, residuals, rank, s = np.linalg.lstsq(X_b, y, rcond=1e-6)
theta_best_svd
```

This function computes $\mathbf{X}^+\mathbf{y}$, where $\mathbf{X}^{+}$ is the _pseudoinverse_ of $\mathbf{X}$ (specifically the Moore-Penrose inverse). You can use `np.linalg.pinv()` to compute the pseudoinverse directly:

```python
np.linalg.pinv(X_b).dot(y)
```

Although the solutions agree, computational complexity isn't the same. 
* naive approach $(X^TX)^{-1}$:  $O(n^3)$
* pseudoinverse $X^+$: $O(n^2)$

Still, if $n\gg 1$ (that is VERY large), we cannot use either.


# 4.2 Gradient Descent

We were lucky that
$\min_{\theta} \frac{1}{m} \sum_i (\theta^Tx^{(i)}-y^{(i)})^2$
has a closed form solution.
Usually, that is not the case.



There's a more general approach to finding the minimum of a cost fuction

$\min_{\theta} f(\theta)$ for arbitrary differentible $f$.

It's called gradient descent (GD).

In GD, you never observe the whole function.

You can only observe the current value $f(\theta)$ and the slope $\frac{df}{d\theta}$. 

For $n>1$ the analogue of the slope is gradient vector $\nabla_{\theta} f=[\frac{df}{d\theta_1},\cdots, \frac{df}{d\theta_n}]^T$.

Picture5.png

Mathematically, 
GD is an iterative update:

Repeat 
$\theta \leftarrow \theta - \eta \frac{d MSE(\theta)}{d \theta}$ (for 1-dim feature)

or

Repeat $\theta \leftarrow \theta - \eta \nabla_{\theta} MSE(\theta)$ (for n-dim features).


Roughly, $\nabla MSE$ determines the direction, and $\eta$ (called learning rate) determines the step size.


Step size does affect the convergence.

Picture6.png

Picture7.png

In this example the cost function $f$ is parabola-shaped as a function of $\theta$. 

As a result, there is a unique minima, which guarrantees the convergence of GD to the solution for apprpopriately chosen $\eta$'s.

MSE of a linear regression, is that function with a parabolic shape. ("convex" function.)

In general, this is not the case.

Lastly, imaging doing GD in two-dimension.

The parabola shape in 2D is a bowl-like function.
The bowl-like function is a convex function (thus GD converges eventually), but the convergence can be slow if the shape is scaled differently for each axis/feature.

## Batch Gradient Descent

The standard GD is also called Batch GD (as opposed to Stochastic GD.)


To use GD, you need to compute the slope (for 1D feature) or the gradient (for n-dim features).
This requires some knowledge of multivariate calculus.


One can show 
$\frac{d MSE(\theta)}{d\theta_j} = \frac{2}{m} \sum_i (\theta^T x^{(i)}-y^{(i)}) x_j^{(i)}$, and therefore compute the gradient

$\nabla_\theta MSE(\theta) = [\frac{d MSE}{d\theta_1}, \cdots, \frac{d MSE}{d\theta_n}] = \frac{2}{m}X^T(X \theta - y).$

Let's implement GD from scratch!!

```python
eta = 0.1  # learning rate
n_iterations = 1000
m = 100

theta = np.random.randn(2,1)  # random initialization

for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients
```

```python
theta_path_bgd = []

def plot_gradient_descent(theta, eta, theta_path=None):
    m = len(X_b)
    plt.plot(X, y, "b.")
    n_iterations = 1000
    for iteration in range(n_iterations):
        if iteration < 10:
            y_predict = X_new_b.dot(theta)
            style = "b-" if iteration > 0 else "r--"
            plt.plot(X_new, y_predict, style)
        gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - eta * gradients
        if theta_path is not None:
            theta_path.append(theta)
    plt.xlabel("$x_1$", fontsize=18)
    #plt.axis([0, 2, 0, 15])
    plt.axis([0,2,-10,20])
    plt.title(r"$\eta = {}$".format(eta), fontsize=16)
```

## Stochastic Gradient Descent

The main problem with Batch Gradient Descent is the fact that it uses the whole training set to compute the gradients at every step:


Repeat $\theta \leftarrow \theta - \eta  \frac{1}{m} \sum_i \nabla [(\theta^Tx^{(i)}-y^{(i)})^2]$.

It makes it very slow when the training set is large.

Q. How do we go around this?

Picture12.png

* Batch GD:

  Repeat $\theta \leftarrow \theta - \eta  \frac{1}{m} \sum_i \nabla [(\theta^Tx^{(i)}-y^{(i)})^2]$.


* Stochastic GD:

  Repeat $\theta \leftarrow \theta - \eta  \nabla [(\theta^Tx^{(i)}-y^{(i)})^2]$ for randomly chosen $i$ each time.


In other words, we move $\theta$ in a direction that is not truely a gradient direction but an approximation of the true gradient direction.

The direction therefore is "noisy" (or "stochastic"), hence the name Stochastic GD.

Since we move $\theta$ along noisy directions, it takes longer to converge.

Q. So what is the advantage?


```python
theta_path_sgd = []
m = len(X_b)
np.random.seed(42)
```

```python
n_epochs = 50
t0, t1 = 5, 50  # learning schedule hyperparameters

def learning_schedule(t):
    return t0 / (t + t1)

theta = np.random.randn(2,1)  # random initialization

for epoch in range(n_epochs):
    for i in range(m):
        if epoch == 0 and i < 20:                    # not shown in the book
            y_predict = X_new_b.dot(theta)           # not shown
            style = "b-" if i > 0 else "r--"         # not shown
            plt.plot(X_new, y_predict, style)        # not shown
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
        theta_path_sgd.append(theta)                 # not shown

plt.plot(X, y, "b.")                                 # not shown
plt.xlabel("$x_1$", fontsize=18)                     # not shown
plt.ylabel("$y$", rotation=0, fontsize=18)           # not shown
plt.axis([0, 2, -10, 20])                              # not shown
save_fig("sgd_plot")                                 # not shown
plt.show()                                           # not shown
```

## Mini-batch gradient descent

```python
theta_path_mgd = []

n_iterations = 50
minibatch_size = 20

np.random.seed(42)
theta = np.random.randn(2,1)  # random initialization

t0, t1 = 200, 1000
def learning_schedule(t):
    return t0 / (t + t1)

t = 0
for epoch in range(n_iterations):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, minibatch_size):
        t += 1
        xi = X_b_shuffled[i:i+minibatch_size]
        yi = y_shuffled[i:i+minibatch_size]
        gradients = 2/minibatch_size * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(t)
        theta = theta - eta * gradients
        theta_path_mgd.append(theta)
```

```python
theta
```

```python
theta_path_bgd = np.array(theta_path_bgd)
theta_path_sgd = np.array(theta_path_sgd)
theta_path_mgd = np.array(theta_path_mgd)
```

```python
plt.figure(figsize=(7,4))
plt.plot(theta_path_sgd[:, 0], theta_path_sgd[:, 1], "r-s", linewidth=1, label="Stochastic")
plt.plot(theta_path_mgd[:, 0], theta_path_mgd[:, 1], "g-+", linewidth=2, label="Mini-batch")
plt.plot(theta_path_bgd[:, 0], theta_path_bgd[:, 1], "b-o", linewidth=3, label="Batch")
plt.legend(loc="upper left", fontsize=16)
plt.xlabel(r"$\theta_0$", fontsize=20)
plt.ylabel(r"$\theta_1$   ", fontsize=20, rotation=0)
plt.axis([2.5, 4.5, 2.3, 3.9])
save_fig("gradient_descent_paths_plot")
plt.show()
```

# Learning Curves

In Lectures 2 and 3, one of the problems was when the model gives of very small training error, but rather large validation error.


Q. What is it called?

Using the fixed training data, we can try fitting it with models of different complexity, and measure training and test error.


Or, similarly, using the same model, we can change the training set size and measure errors.

Let's try it with a linear regression.

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=10)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train) + 1):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))

    plt.plot(np.sqrt(train_errors), "r+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b.", linewidth=3, label="val")
    plt.legend(loc="upper right", fontsize=14)   # not shown in the book
    plt.xlabel("Training set size", fontsize=14) # not shown
    plt.ylabel("RMSE", fontsize=14)              # not shown
```

```python
lin_reg = LinearRegression()
plot_learning_curves(lin_reg, X, y)
plt.axis([0, 80, 0, 3])                         # not shown in the book
save_fig("underfitting_learning_curves_plot")   # not shown
plt.show()                                      # not shown
```

# 4.5 Regularized Linear Models

Conclusions from above:
* More complex models are ultimately better if you have enough training data
* But if you don't have enough training data, overfitting happens and therefore simpler models are better.



In high-dimensions or with noise in the data, even Linear models can overfit !!!




The most popular method to fight overfitting is **regularization**.

Regularizated models have additional terms in their cost function other than the original cost (like MSE).

There are many choices of regularization functions resulting in many variations of the original model.

For exmaple, the simple linear regression can modified by adding various types of regularization functions, yielding

Ridge Regression, Lasso, Elastic Net, etc.

```python
from sklearn.linear_model import Ridge

def plot_model(model_class, polynomial, alphas, **model_kargs):
    for alpha, style in zip(alphas, ("b-", "g--", "r:")):
        model = model_class(alpha, **model_kargs) if alpha > 0 else LinearRegression()
        if polynomial:
            model = Pipeline([
                    ("poly_features", PolynomialFeatures(degree=10, include_bias=False)),
                    ("std_scaler", StandardScaler()),
                    ("regul_reg", model),
                ])
        model.fit(X, y)
        y_new_regul = model.predict(X_new)
        lw = 2 if alpha > 0 else 1
        plt.plot(X_new, y_new_regul, style, linewidth=lw, label=r"$\alpha = {}$".format(alpha))
    plt.plot(X, y, "b.", linewidth=3)
    plt.legend(loc="upper left", fontsize=15)
    plt.xlabel("$x_1$", fontsize=18)
    plt.axis([0, 3, 0, 4])

plt.figure(figsize=(8,4))
plt.subplot(121)
plot_model(Ridge, polynomial=False, alphas=(0, 10, 100), random_state=42)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.subplot(122)
plot_model(Ridge, polynomial=True, alphas=(0, 10**-5, 1), random_state=42)

save_fig("ridge_regression_plot")
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

t1a, t1b, t2a, t2b = -1, 3, -1.5, 1.5

t1s = np.linspace(t1a, t1b, 500)
t2s = np.linspace(t2a, t2b, 500)
t1, t2 = np.meshgrid(t1s, t2s)
T = np.c_[t1.ravel(), t2.ravel()]
Xr = np.array([[1, 1], [1, -1], [1, 0.5]])
yr = 2 * Xr[:, :1] + 0.5 * Xr[:, 1:]

J = (1/len(Xr) * np.sum((T.dot(Xr.T) - yr.T)**2, axis=1)).reshape(t1.shape)

N1 = np.linalg.norm(T, ord=1, axis=1).reshape(t1.shape)
N2 = np.linalg.norm(T, ord=2, axis=1).reshape(t1.shape)

t_min_idx = np.unravel_index(np.argmin(J), J.shape)
t1_min, t2_min = t1[t_min_idx], t2[t_min_idx]

t_init = np.array([[0.25], [-1]])


def bgd_path(theta, X, y, l1, l2, core = 1, eta = 0.05, n_iterations = 200):
    path = [theta]
    for iteration in range(n_iterations):
        gradients = core * 2/len(X) * X.T.dot(X.dot(theta) - y) + l1 * np.sign(theta) + l2 * theta
        theta = theta - eta * gradients
        path.append(theta)
    return np.array(path)

fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(10.1, 8))
for i, N, l1, l2, title in ((0, N1, 2., 0, "Lasso"), (1, N2, 0,  2., "Ridge")):
    JR = J + l1 * N1 + l2 * 0.5 * N2**2
    
    tr_min_idx = np.unravel_index(np.argmin(JR), JR.shape)
    t1r_min, t2r_min = t1[tr_min_idx], t2[tr_min_idx]

    levelsJ=(np.exp(np.linspace(0, 1, 20)) - 1) * (np.max(J) - np.min(J)) + np.min(J)
    levelsJR=(np.exp(np.linspace(0, 1, 20)) - 1) * (np.max(JR) - np.min(JR)) + np.min(JR)
    levelsN=np.linspace(0, np.max(N), 10)
    
    path_J = bgd_path(t_init, Xr, yr, l1=0, l2=0)
    path_JR = bgd_path(t_init, Xr, yr, l1, l2)
    path_N = bgd_path(np.array([[2.0], [0.5]]), Xr, yr, np.sign(l1)/3, np.sign(l2), core=0)

    ax = axes[i, 0]
    ax.grid(True)
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    ax.contourf(t1, t2, N / 2., levels=levelsN)
    ax.plot(path_N[:, 0], path_N[:, 1], "y--")
    ax.plot(0, 0, "ys")
    ax.plot(t1_min, t2_min, "ys")
    ax.set_title(r"$\ell_{}$ penalty".format(i + 1), fontsize=16)
    ax.axis([t1a, t1b, t2a, t2b])
    if i == 1:
        ax.set_xlabel(r"$\theta_1$", fontsize=16)
    ax.set_ylabel(r"$\theta_2$", fontsize=16, rotation=0)

    ax = axes[i, 1]
    ax.grid(True)
    ax.axhline(y=0, color='k')
    ax.axvline(x=0, color='k')
    ax.contourf(t1, t2, JR, levels=levelsJR, alpha=0.9)
    ax.plot(path_JR[:, 0], path_JR[:, 1], "w-o")
    ax.plot(path_N[:, 0], path_N[:, 1], "y--")
    ax.plot(0, 0, "ys")
    ax.plot(t1_min, t2_min, "ys")
    ax.plot(t1r_min, t2r_min, "rs")
    ax.set_title(title, fontsize=16)
    ax.axis([t1a, t1b, t2a, t2b])
    if i == 1:
        ax.set_xlabel(r"$\theta_1$", fontsize=16)

save_fig("lasso_vs_ridge_plot")
plt.show()
```

## Estimating Probabilities

We've seen several linear classifiers already. 

Here, let's look at a different kind of classifier called logistic regression. (Why is it called regression?)

A classifier typically outputs catergories: 0, 1, ...

What if a classify not only outputs this, but also how probable/confident it is about the prediction?

For example, output=$[0: 15\%, 1: 75\%, 2: 10\%]$.

Q. When will such information be useful? Appalachian trail ...

For simplicity, assume a binary classification problem ($\mathcal{Y}=\{-1,+1\}$).

The linear decision function was:

* y_pred(x) = 1, if $\theta^Tx \geq 0$, and
* y_pred(x) = -1, if $\theta^Tx < 0$.

In addition, we want to predict $Pr[y=1|x]$.

* $Pr[y=1|x]$: probability/confidence that the classifier thinks the input $x$ is of class 1.

Note that $Pr[y=1|x] = 1 - Pr[y=-1|x]$. 

The $Pr[\cdot]$ is non-negative and between 0 and 1.

One of the simplest way to produce the probability with a linear function is:

* $Pr[y=1|x] = \sigma(\theta^Tx)$, where

  * $\sigma(t) \to 1\;\;\textrm{if}\;\;t \to \infty$ and
  * $\sigma(t) \to 0\;\;\textrm{if}\;\;t \to -\infty$.
  

The $\sigma(\cdot)$ is called the logistic/sigmoid function:

* $\sigma(t):= \frac{1}{1+e^{-t}}$

(Therefore, $Pr[y=1|x]=\frac{1}{1+e^{-\theta^T x}}$),

```python
t = np.linspace(-10, 10, 100)
sig = 1 / (1 + np.exp(-t))
plt.figure(figsize=(9, 3))
plt.plot([-10, 10], [0, 0], "k-")
plt.plot([-10, 10], [0.5, 0.5], "k:")
plt.plot([-10, 10], [1, 1], "k:")
plt.plot([0, 0], [-1.1, 1.1], "k-")
plt.plot(t, sig, "b-", linewidth=2, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$")
plt.xlabel("t")
plt.legend(loc="upper left", fontsize=20)
plt.axis([-10, 10, -0.1, 1.1])
save_fig("logistic_function_plot")
plt.show()
```

Q. What is the parameter of the logistic regression model? 

Q. How do we find the parameters?

Cost $f(\theta)=\frac{1}{m}\sum_i \left[-y^{(i)} \log \sigma(\theta^Tx) -(1-y^{(i)})\log (1-\sigma(\theta^Tx))\right]$


Q. You can't find the optimal $\theta$ in closed-form. What do you do?

Hint:
$\frac{\partial f}{\partial \theta_j} = \frac{1}{m}\sum_i (\sigma(\theta^T x^{(i)})-y^{(i)})\; x_j^{(i)}$.

## Softmax Regression

Q. Logistic regression is inherently a binary classifier. 
What did we do last time to create a multiclass classifier out of binary classifier?


However, there is a direct extension of logistic regression for multiclass problems. 

It's called softmax regression. As with any model, we have to know two things:
  * How does it makes predictions? (That is, what is the decision function?)
  * What's the cost function (and how to minimize it)?


* Prediction:
y_pred(x) = $\arg\max_k\;Pr[y=k|x]=\frac{ e^{(\theta^{(k)})^Tx} }{\sum_j e^{(\theta^{(j)})^Tx}}$.

Parameters: $ \theta^{(1)}, \theta^{(1)}, ... , \theta^{(K)}$.


* Cost function:
$f(\theta) = -\frac{1}{m} \sum_i \sum_k y_k^{(i)} \log Pr[y=k|x^{(i)}]$





Details are given in the optional section below.

## Softmax regression in detail


The softmax function is defined as 
$\sigma\left(\mathbf{s}(\mathbf{x})\right)_k = \dfrac{\exp\left(s_k(\mathbf{x})\right)}{\sum\limits_{j=1}^{K}{\exp\left(s_j(\mathbf{x})\right)}}$

Let's implement the Softmax function. 

```python
def softmax(logits):
    exps = np.exp(logits)
    exp_sums = np.sum(exps, axis=1, keepdims=True)
    return exps / exp_sums
```

We are almost ready to start training. Let's define the number of inputs and outputs:

```python
n_inputs = X_train.shape[1] # == 3 (2 features plus the bias term)
n_outputs = len(np.unique(y_train))   # == 3 (3 iris classes)
```

Now here comes the hardest part: training! Theoretically, it's simple: it's just a matter of translating the math equations into Python code. But in practice, it can be quite tricky: in particular, it's easy to mix up the order of the terms, or the indices. You can even end up with code that looks like it's working but is actually not computing exactly the right thing. When unsure, you should write down the shape of each term in the equation and make sure the corresponding terms in your code match closely. It can also help to evaluate each term independently and print them out. The good news it that you won't have to do this everyday, since all this is well implemented by Scikit-Learn, but it will help you understand what's going on under the hood.

So the equations we will need are the cost function:

$J(\mathbf{\Theta}) =
- \dfrac{1}{m}\sum\limits_{i=1}^{m}\sum\limits_{k=1}^{K}{y_k^{(i)}\log\left(\hat{p}_k^{(i)}\right)}$

And the equation for the gradients:

$\nabla_{\mathbf{\theta}^{(k)}} \, J(\mathbf{\Theta}) = \dfrac{1}{m} \sum\limits_{i=1}^{m}{ \left ( \hat{p}^{(i)}_k - y_k^{(i)} \right ) \mathbf{x}^{(i)}}$

Note that $\log\left(\hat{p}_k^{(i)}\right)$ may not be computable if $\hat{p}_k^{(i)} = 0$. So we will add a tiny value $\epsilon$ to $\log\left(\hat{p}_k^{(i)}\right)$ to avoid getting `nan` values.

```python
eta = 0.01
n_iterations = 5001
m = len(X_train)
epsilon = 1e-7

Theta = np.random.randn(n_inputs, n_outputs)

for iteration in range(n_iterations):
    logits = X_train.dot(Theta)
    Y_proba = softmax(logits)
    if iteration % 500 == 0:
        loss = -np.mean(np.sum(Y_train_one_hot * np.log(Y_proba + epsilon), axis=1))
        print(iteration, loss)
    error = Y_proba - Y_train_one_hot
    gradients = 1/m * X_train.T.dot(error)
    Theta = Theta - eta * gradients
```

Well, this model looks pretty good. For the sake of the exercise, let's add a bit of $\ell_2$ regularization. The following training code is similar to the one above, but the loss now has an additional $\ell_2$ penalty, and the gradients have the proper additional term (note that we don't regularize the first element of `Theta` since this corresponds to the bias term). Also, let's try increasing the learning rate `eta`.

```python
eta = 0.1
n_iterations = 5001
m = len(X_train)
epsilon = 1e-7
alpha = 0.1  # regularization hyperparameter

Theta = np.random.randn(n_inputs, n_outputs)

for iteration in range(n_iterations):
    logits = X_train.dot(Theta)
    Y_proba = softmax(logits)
    if iteration % 500 == 0:
        xentropy_loss = -np.mean(np.sum(Y_train_one_hot * np.log(Y_proba + epsilon), axis=1))
        l2_loss = 1/2 * np.sum(np.square(Theta[1:]))
        loss = xentropy_loss + alpha * l2_loss
        print(iteration, loss)
    error = Y_proba - Y_train_one_hot
    gradients = 1/m * X_train.T.dot(error) + np.r_[np.zeros([1, n_outputs]), alpha * Theta[1:]]
    Theta = Theta - eta * gradients
```

Now let's add early stopping. For this we just need to measure the loss on the validation set at every iteration and stop when the error starts growing.

```python
eta = 0.1 
n_iterations = 5001
m = len(X_train)
epsilon = 1e-7
alpha = 0.1  # regularization hyperparameter
best_loss = np.infty

Theta = np.random.randn(n_inputs, n_outputs)

for iteration in range(n_iterations):
    logits = X_train.dot(Theta)
    Y_proba = softmax(logits)
    error = Y_proba - Y_train_one_hot
    gradients = 1/m * X_train.T.dot(error) + np.r_[np.zeros([1, n_outputs]), alpha * Theta[1:]]
    Theta = Theta - eta * gradients

    logits = X_valid.dot(Theta)
    Y_proba = softmax(logits)
    xentropy_loss = -np.mean(np.sum(Y_valid_one_hot * np.log(Y_proba + epsilon), axis=1))
    l2_loss = 1/2 * np.sum(np.square(Theta[1:]))
    loss = xentropy_loss + alpha * l2_loss
    if iteration % 500 == 0:
        print(iteration, loss)
    if loss < best_loss:
        best_loss = loss
    else:
        print(iteration - 1, best_loss)
        print(iteration, loss, "early stopping!")
        break
```
